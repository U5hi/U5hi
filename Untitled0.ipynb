```python
import random

angka_rahasia = random.randint (1, 50)
kesempatan = 3

while kesempatan > 0:
    tebakan = int(input("Tebak angka dari 1 sampai 50:"))
    if tebakan == angka_rahasia:
      print("Hebat banget ! kamu benar")
      break
    else:
      kesempatan -= 1
      print("Salah. masih ada kesempatan tersisa:", kesempatan)
if kesempatan == 0:
     print("Yah, habis kesempatan. Angka rahasianya adalah", angka_rahasia)
```

```python
import numpy as np
import matplotlib.pyplot as plt

kelopak = 10
theta = np.linspace(0, 2 * np.pi, 1000)
r = np.sin (kelopak * theta)

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(6, 6))
ax.plot(theta, r, color='blue' )

ax.set_facecolor('lightblue')
ax.set_thetagirds([])
ax.spines['polar'].set_visible(false)
ax.set_yticklables([])

ax.set_title(f'Grafik Bunga dengan {kelopak} kelopak', fontsize=14)
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt

kelopak = 10
theta = np.linspace(0, 2 * np.pi, 1000)
r = np.sin (kelopak * theta)

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(6, 6))
ax.plot(theta, r, color='darkblue' )

ax.set_facecolor('silver')
# Correct method to hide theta grid lines in polar plots
ax.grid(False, which='both', axis='x')
# Correct spelling for hiding y-tick labels
ax.set_yticklabels([])
# Correct boolean value for set_visible
ax.spines['polar'].set_visible(False)


ax.set_title(f'Grafik Bunga dengan {kelopak} kelopak', fontsize=14)
# Call the function to display the plot
plt.show()
```
